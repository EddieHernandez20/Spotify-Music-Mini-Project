```python
#Recode Mood to numbers
#Recode for Happy
def mood1 (series):
    if series == "Happy":
        return 1
    if series == "Sad":
        return 2
    if series == "Energetic":
        return 3
    if series == "Calm":
        return 4
```

```python
musc1['MoodN'] = musc1['mood'].apply(mood1)
```
